TASK2 Unemployment analysis with Python

Unemployment is measured by the unemployment rate which is the number of people
who are unemployed as a percentage of the total labour force. We have seen a sharp
increase in the unemployment rate during Covid-19, so analyzing the unemployment rate
can be a good data science project.




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv('Unemployment in India.csv')
data
```

```python
#droping the null values
data.dropna(axis=0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
sns.pairplot(data, hue="Region")
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data.corr(),annot=True)
```

```python
data.columns=["Region",	"Date","Frequency",	"Estimated Unemployment Rate (%)","Estimated Employed","Estimated Labour Participation Rate (%)",	"Area"]
plt.title("Indian Unemployment")
sns.histplot(data=data, x="Estimated Employed", hue='Region')
```

```python
plt.figure(figsize=(18,14))
plt.title("Indian Unemployment")
sns.histplot(data=data, x="Estimated Unemployment Rate (%)", hue='Region')
```

```python
unemployment = data[["Region", "Estimated Unemployment Rate (%)"]]
figure = px.sunburst(unemployment, path=["Region"],values="Estimated Unemployment Rate (%)",hover_data=["Estimated Unemployment Rate (%)"],width=900,height=900,color_continuous_scale="RdY2Gn",title="Unemployment Rate in India")
figure.update_traces(hovertemplate='%{label}: %{value:.1f}%')
```
